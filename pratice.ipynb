```python
#palindrome or not
a=121
n=a
s=0
while a>0:
  rev=a%10
  s=(s*10)+rev
  a=a//10
if n==s:
  print("given number is palindrome")
else:
  print("given number is not palindrome")
```

```python
# armstrong or not
a=153
n=a
s=0
while a>0:
  rev=a%10
  s=s+(rev*rev*rev)
  a=a//10
if n==s:
  print("given number is armstrong")
else:
  print("given number is not armstrong")
```

```python
#perfect or not
n=6
s=0
for i in range(1,n):
  if n%i==0:
    s=s+i
if s==n:
  print("perfect number")
else:
  print("not perfect number")

```

```python
#Strong or not
n=145
t=n
s=0
while n>0:
  rev=n%10
  f=1
  for i in range(1,rev+1):
    f=f*i
  s=s+f
  n=n//10
if t==s:
  print("Strong number")
else:
  print("not strong")
```

```python
#prime or not
a=3
c=0
for i in range(1,a+1):
  if a%i==0:
    c=c+1
if c==2:
  print("prime")
else:
  print("not prime")
```

```python
#prime between 2 to 100
for i in range(2,101):
  c=0
  for j in range(1,i+1):
    if i%j==0:
      c=c+1
  if c==2:
    print(f"{i} is prime")
  else:
    print(f"{i} is not prime")
```

```python
#fibnocaci series
n=8
a=0
b=1
print(a)
print(b)
for i in range(0,n-2):
  c=a+b
  print(c)
  a=b
  b=c
```
