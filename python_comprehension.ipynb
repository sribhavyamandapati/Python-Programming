**dictionary**

```python
fruits={"apple":100,"banana":200,"carrot":300}
for i in fruits:
  print(fruits[i]+100)
```

```python
fruit={"apple":100,"banana":200,"carrot":300}
for i in fruit:
  if fruit[i]>150:
    print(fruit[i]*2)
```

```python
print([fruit[i]*2 for i in fruit if fruit[i]>150])
```

```python
fruit={"apple":100,"banana":200,"carrot":300}
for i in fruit:
  if fruit[i]>150:
    print(fruit[i]*2)
  else:
    print(fruit[i]*3)
```

```python
print([fruit[i]*2 if fruit[i]>150  else fruit[i]*3 for i in fruit])
```

```python
even=[]
for num in range(1,11):
  if num%2==0:
    even.append(num)
print(even)
```

```python
even=[ num for num in range(1,11) if num%2==0]
print(even)
```

```python
result=[]
for i in range(1,11):
  if i%2==0:
    result.append(i**2)
  else:
    result.append(i**3)
print(result)
```

```python
result=[i**2  if i%2==0 else  i**3 for i in range(1,11)]
print(result)
```

```python
list1=[1,2,3,4]
list2=[1,20,30,4]
result=[]
for x in list1:
  for y in list2:
    if x==y:
      result.append(y)
print(result)
```
